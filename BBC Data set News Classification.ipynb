#### Feature extraction and model training

```python
import re
from sklearn.feature_extraction.text import TfidfVectorizer
from textblob import Word
import numpy as np
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import MultinomialNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, cohen_kappa_score, confusion_matrix, classification_report
```

```python
def clean_str(string):
    """
    Tokenization/string cleaning for datasets.
    Original taken from https://github.com/yoonkim/CNN_sentence/blob/master/process_data.py
    """
    string = re.sub(r"\'s", "", string)
    string = re.sub(r"\'ve", "", string)
    string = re.sub(r"n\'t", "", string)
    string = re.sub(r"\'re", "", string)
    string = re.sub(r"\'d", "", string)
    string = re.sub(r"\'ll", "", string)
    string = re.sub(r",", "", string)
    string = re.sub(r"!", " ! ", string)
    string = re.sub(r"\(", "", string)
    string = re.sub(r"\)", "", string)
    string = re.sub(r"\?", "", string)
    string = re.sub(r"'", "", string)
    string = re.sub(r"[^A-Za-z0-9(),!?\'\`]", " ", string)
    string = re.sub(r"[0-9]\w+|[0-9]","", string)
    string = re.sub(r"\s{2,}", " ", string)
    return string.strip().lower()

data = pd.read_csv('/Users/Admin/Desktop/Python Codes/BBC/dataset.csv')
x = data['news'].tolist()
y = data['type'].tolist()

for index,value in enumerate(x):
    print("processing data:",index)
    x[index] = ' '.join([Word(word).lemmatize() for word in clean_str(value).split()])
```

```python
vect = TfidfVectorizer(stop_words='english',min_df=2)
```

```python
X = vect.fit_transform(x)
Y = np.array(y)

print("no of features extracted:",X.shape[1])
```

#### Training the model - Data set split into training set - 1780 examples + test set - 425 examples

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.20, random_state=42)

print("train size:", X_train.shape)
print("test size:", X_test.shape)
```

```python
# Naive Bayes classifier
nb_model = MultinomialNB().fit(X_train, y_train)

# Decision tree classifier
dtree = DecisionTreeClassifier()
dtree.fit(X_train, y_train)

# Random forest classifier
rfc_model = RandomForestClassifier(n_estimators=300, max_depth=150,n_jobs=1)
rfc_model.fit(X_train, y_train)
```

#### Model Evaluation

```python
# Naive Bayes Evaluation
y_pred_nb = nb_model.predict(X_test)
c_mat = confusion_matrix(y_test,y_pred_nb)
c_report = classification_report(y_test, y_pred_nb)
kappa = cohen_kappa_score(y_test,y_pred_nb)
acc = accuracy_score(y_test,y_pred_nb)
print("NAIVE BAYES \n")
print("Confusion Matrix:\n", c_mat)
print ("\nClassification_report : \n", c_report)
print("\nKappa: ",kappa)
print("\nAccuracy: ",acc)


# Decision Tree Evaluation
y_pred_dt = dtree.predict(X_test)
c_mat_dt = confusion_matrix(y_test,y_pred_dt)
c_report_dt = classification_report(y_test, y_pred_dt)
kappa_dt = cohen_kappa_score(y_test,y_pred_dt)
acc_dt = accuracy_score(y_test,y_pred_dt)
print("\n\nDECISION TREE \n")
print("Confusion Matrix:\n", c_mat_dt)
print ("\nClassification_report : \n", c_report_dt)
print("\nKappa: ",kappa_dt)
print("\nAccuracy: ",acc_dt)


# Random Forest Evaluation
y_pred_rfc = rfc_model.predict(X_test)
c_mat_rfc = confusion_matrix(y_test,y_pred_rfc)
c_report_rfc = classification_report(y_test, y_pred_rfc)
kappa_rfc = cohen_kappa_score(y_test,y_pred_rfc)
acc_rfc = accuracy_score(y_test,y_pred_rfc)
print("\n\nRANDOM FOREST \n")
print("Confusion Matrix:\n", c_mat_rfc)
print ("\nClassification_report : \n", c_report_rfc)
print("\nKappa: ",kappa_rfc)
print("\nAccuracy: ",acc_rfc)




```
